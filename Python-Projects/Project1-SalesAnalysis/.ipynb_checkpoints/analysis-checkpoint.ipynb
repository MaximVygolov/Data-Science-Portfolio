```python
# Загрузка необходимых библиотек
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import seaborn as sns
```

```python
# Загрузка данных
file_path = "data/US_Regional_Sales_Data.csv"
df = pd.read_csv(file_path)
```

```python
df.columns
```

```python
df.info()
```

```python
# Преобразование столбцов дат в правильный формат
df['ProcuredDate'] = pd.to_datetime(df['ProcuredDate'], format='%d/%m/%y', errors='coerce')
df['OrderDate'] = pd.to_datetime(df['OrderDate'], format='%d/%m/%y', errors='coerce')
df['ShipDate'] = pd.to_datetime(df['ShipDate'], format='%d/%m/%y', errors='coerce')
df['DeliveryDate'] = pd.to_datetime(df['DeliveryDate'], format='%d/%m/%y', errors='coerce')
```

```python
# Преобразование числовых значений из строк в числа
df['Unit Cost'] = df['Unit Cost'].str.replace(',', '').astype(float)
df['Unit Price'] = df['Unit Price'].str.replace(',', '').astype(float)
```

```python
df.head()
```

```python
# Удаление ненужных столбцов для анализа
columns_to_drop = ['OrderNumber', 'CurrencyCode', '_SalesTeamID', '_CustomerID', '_StoreID', '_ProductID']
df.drop(columns=columns_to_drop, inplace=True)
```

```python
# df.head()
```

```python
# Добавление столбца с годом и месяцем для анализа по времени
df['YearMonth'] = df['OrderDate'].dt.to_period('M')

# Добавление столбца с общим объемом продаж
df['Total Sales'] = df['Order Quantity'] * df['Unit Price']
```

```python
# df.head()
```

```python
# Анализ объемов продаж по регионам
region_sales = df.groupby('WarehouseCode')['Total Sales'].sum().reset_index()

# Визуализация объемов продаж по регионам
plt.figure(figsize=(10, 6))
sns.barplot(x='WarehouseCode', y='Total Sales', data=region_sales, palette='viridis')
plt.title('Total Sales by Region')
plt.xlabel('Warehouse Code')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()

# Сохранение графика
plt.savefig('visualizations/total_sales_by_region.png')

plt.show()
```

```python
# Анализ продаж по времени
time_sales = df.groupby('YearMonth')['Total Sales'].sum().reset_index()
time_sales['YearMonth'] = time_sales['YearMonth'].astype(str)

plt.figure(figsize=(12, 6))
sns.lineplot(x='YearMonth', y='Total Sales', data=time_sales, marker='o')
plt.title('Sales Over Time')
plt.xlabel('Year-Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()

# Сохранение графика
plt.savefig('visualizations/sales_over_time.png')
plt.show()
```

```python
# Анализ объемов по каналам продаж

#Получение цветов из палитры viridis
num_channels = df['Sales Channel'].nunique()
colors = cm.viridis([i / num_channels for i in range(num_channels)])

# Построение круговой диаграммы
plt.figure(figsize=(10, 6))
df.groupby('Sales Channel')['Order Quantity'].sum().plot(
    kind='pie',
    autopct='%0.2f%%',
    colors=colors,
    textprops={'color': 'white'},
    title='Quantity of Orders by Sales Channels'
)

# Добавление легенды
plt.legend(df['Sales Channel'].unique(), title="Sales Channel", bbox_to_anchor=(1.1, 0.8), loc='center left')
plt.ylabel('')

# Сохранение графика
plt.savefig('visualizations/quantity_of_orders_by_sales_channels.png')

plt.show()
```
