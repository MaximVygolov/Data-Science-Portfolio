### **Описание проекта: Анализ и прогнозирование оттока клиентов телекоммуникационной компании**

#### **Краткое описание**
Этот проект посвящен анализу данных телекоммуникационной компании и прогнозированию оттока клиентов. Основная цель — выявить ключевые факторы, влияющие на отток, и построить модель, которая сможет прогнозировать, уйдёт клиент или нет. Проект также включает визуализацию данных, проверку гипотез и сравнение моделей машинного обучения.

---

#### **Контекст задачи**
Каждая компания стремится не только привлекать новых клиентов, но и удерживать существующих, так как привлечение новых клиентов обходится дороже. Кроме того, новые клиенты могут быть менее лояльны или не видеть ценности в услугах компании. Удержание существующих клиентов — более экономичный способ поддержания стабильного дохода.

Прогнозирование оттока позволяет:
- **Выявлять клиентов, которые находятся в зоне риска.**
- **Разрабатывать стратегии удержания**, предлагая им персонализированные услуги или скидки.
- **Снижать затраты на привлечение новых клиентов**, одновременно поддерживая текущую базу.

### Гипотезы для исследования

Ниже приведены пять гипотез, основанные на предоставленных данных и контексте задачи. Эти гипотезы помогут в анализе данных и построении модели прогнозирования оттока пользователей.

#### Гипотеза 1:

Новые клиенты с высокой стоимостью ежемесячной оплаты **(MonthlyCharges)** чаще склонны к оттоку.<br>
***Обоснование:*** Короткий срок использования услуг (tenure) и высокая стоимость могут быть основными факторами, вызывающими недовольство клиентов, так как они еще не успели привыкнуть к услугам компании и не видят ценности за высокую стоимость.

#### Гипотеза 2:

Клиенты, использующие интернет через **Fiber optic**, более склонны к оттоку по сравнению с пользователями **DSL**.<br>
***Обоснование:*** Несмотря на высокую скорость, стоимость услуг **Fiber optic** выше, и это может быть фактором неудовлетворенности.

#### Гипотеза 3:

Клиенты, у которых контракт **Month-to-month**, чаще уходят, чем клиенты с долгосрочными контрактами (One year или Two year).<br>
***Обоснование:*** Краткосрочные контракты легче расторгнуть, что делает клиентов с этим типом контракта более склонными к оттоку.

#### Гипотеза 4:

Клиенты, которые не используют дополнительные услуги, такие как **TechSupport** или **OnlineBackup**, чаще склонны к оттоку.<br>
***Обоснование:*** Отсутствие дополнительных услуг может свидетельствовать о низкой вовлеченности клиентов или недостатке ценности, которую они получают от оператора.

#### Гипотеза 5:

Долгосрочные клиенты **(с высоким значением tenure)** менее склонны к оттоку.<br>
***Обоснование:*** Долгосрочные клиенты, вероятно, уже встроены в экосистему услуг и привыкли к их использованию, что снижает вероятность оттока.

## Проверка гипотезы № 1 (графически и с помощью модели)

```python
# График распределения MonthlyCharges относительно Churn и tenure
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='tenure', y='MonthlyCharges', hue='Churn', alpha=0.7)
plt.title('Взаимосвязь MonthlyCharges, tenure и Churn')
plt.xlabel('Tenure (месяцы)')
plt.ylabel('MonthlyCharges (USD)')
plt.legend(title='Churn')
plt.show()

# Дополнительно: Разделение на группы по tenure
df['tenure_group'] = pd.cut(df['tenure'], bins=[0, 12, 24, 36, 48, 60, 72], labels=['0-12', '12-24', '24-36', '36-48', '48-60', '60-72'])

plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='tenure_group', y='MonthlyCharges', hue='Churn')
plt.title('Распределение MonthlyCharges по группам tenure и Churn')
plt.xlabel('Tenure Group (месяцы)')
plt.ylabel('MonthlyCharges (USD)')
plt.legend(title='Churn')
plt.show()
```

- По мере увеличения `tenure` (особенно после 36 месяцев) становится заметным, что большинство клиентов остаются (`Churn = No`), независимо от уровня `MonthlyCharges`.
- Лояльность пользователей растет с увеличением срока их сотрудничества, даже при высоких ежемесячных расходах.

### 2. Проверка зависимости с помощью модели

Теперь мы добавим оба признака (MonthlyCharges и tenure) в модель логистической регрессии.

```python
# Подготовка данных
X = df[['MonthlyCharges', 'tenure']]  # Используем два признака
y = df['Churn'].apply(lambda x: 1 if x == 'Yes' else 0)  # Преобразуем Churn в 0 и 1

# Разделение на обучающую и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Обучение модели логистической регрессии
model = LogisticRegression(class_weight='balanced', random_state=42)
model.fit(X_train, y_train)

# Прогнозирование
y_pred = model.predict(X_test)
y_proba = model.predict_proba(X_test)[:, 1]

# Оценка модели
print("Качество модели Logistic Regression:")
print(classification_report(y_test, y_pred))
print(f"AUC-ROC: {roc_auc_score(y_test, y_proba):.2f}")
```

### **Выводы после модели Logistic Regression**

#### **Метрики модели:**
- **Accuracy (точность модели):** 72% — модель корректно классифицирует 72% пользователей.
- **Precision (точность предсказания класса `1` — Churn):** 47% — из всех предсказанных "ушедших клиентов" только 47% действительно ушли.
- **Recall (полнота для класса `1` — Churn):** 70% — модель правильно определила 70% из всех реальных случаев оттока.
- **F1-score для класса `1`:** 56% — сбалансированный показатель между точностью и полнотой для оттока.
- **AUC-ROC:** 0.79 — модель демонстрирует хорошее качество при разделении классов и превышает базовый уровень (0.5).

#### **Анализ результатов:**
1. **Сильные стороны модели:**
   - Модель хорошо распознает клиентов, склонных к оттоку (`Recall = 70%`).
   - AUC-ROC = 0.79 указывает на то, что модель способна эффективно разделять классы (`Churn` и `No Churn`).
<br><br>
2. **Слабые стороны модели:**
   - Низкое значение Precision (`47%`) для класса `1` говорит о большом количестве ложноположительных предсказаний (клиенты, которые ошибочно классифицированы как склонные к оттоку).
   - Общая точность (Accuracy = 72%) могла бы быть выше при использовании дополнительных признаков.
<br><br>
3. **Вклад признаков:**
   - Добавление `tenure` к `MonthlyCharges` улучшило качество модели, о чем свидетельствует AUC-ROC (0.79). Это подтверждает, что комбинация этих признаков полезна для прогнозирования оттока.

# 3. Построение моделей для прогнозирования оттока

В этом разделе мы будем:

1. Подготавливать данные на основе проверенных гипотез.
2. Обучать несколько моделей машинного обучения.
3. Сравнивать их качество и обосновывать выбор лучшей модели.

На основе гипотез и выявленных взаимосвязей мы выбираем следующие признаки для построения модели:

- tenure (время использования услуг).
- MonthlyCharges (ежемесячные расходы).
- Contract (тип контракта).
- Дополнительные услуги: TechSupport, OnlineBackup, DeviceProtection.
- InternetService (тип интернет-сервиса).

```python
# Выбор признаков
features = ['tenure', 'MonthlyCharges', 'Contract', 'InternetService', 'TechSupport', 'OnlineBackup', 'DeviceProtection']
X = df[features]
y = df['Churn'].apply(lambda x: 1 if x == 'Yes' else 0)  # Целевая переменная

# Преобразование категориальных признаков в числовые (One-Hot Encoding)
encoder = OneHotEncoder(drop='first', sparse_output=False)
X_encoded = encoder.fit_transform(X.select_dtypes(include='object'))

# Объединение закодированных категориальных признаков с числовыми
X_final = pd.concat([pd.DataFrame(X_encoded, columns=encoder.get_feature_names_out()), X.select_dtypes(exclude='object')], axis=1)

# Разделение на обучающую и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(X_final, y, test_size=0.3, random_state=42)
```

Мы обучим несколько моделей и сравним их результаты:

- Logistic Regression.
- Random Forest.
- Gradient Boosting (например, с помощью XGBoost).

```python
# Логистическая регрессия
log_model = LogisticRegression(class_weight='balanced', random_state=42)
log_model.fit(X_train, y_train)

# Прогнозы
y_pred_log = log_model.predict(X_test)
y_proba_log = log_model.predict_proba(X_test)[:, 1]

# Оценка качества
print("Метрики модели Logistic Regression:")
print(classification_report(y_test, y_pred_log))
print(f"AUC-ROC: {roc_auc_score(y_test, y_proba_log):.2f}")
```

```python
# Случайный лес
rf_model = RandomForestClassifier(class_weight='balanced', random_state=42)
rf_model.fit(X_train, y_train)

# Прогнозы
y_pred_rf = rf_model.predict(X_test)
y_proba_rf = rf_model.predict_proba(X_test)[:, 1]

# Оценка качества
print("Метрики модели Random Forest:")
print(classification_report(y_test, y_pred_rf))
print(f"AUC-ROC: {roc_auc_score(y_test, y_proba_rf):.2f}")
```

```python
# Gradient Boosting
xgb_model = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
xgb_model.fit(X_train, y_train)

# Прогнозы
y_pred_xgb = xgb_model.predict(X_test)
y_proba_xgb = xgb_model.predict_proba(X_test)[:, 1]

# Оценка качества
print("Метрики модели Gradient Boosting (XGBoost):")
print(classification_report(y_test, y_pred_xgb))
print(f"AUC-ROC: {roc_auc_score(y_test, y_proba_xgb):.2f}")
```

# 4. Сравнение качества моделей

Создадим таблицу для сравнения метрик Accuracy, Recall, Precision, F1-Score и AUC-ROC для каждой модели.

```python
# Сравнение метрик
comparison = pd.DataFrame({
    'Model': ['Logistic Regression', 'Random Forest', 'Gradient Boosting'],
    'AUC-ROC': [roc_auc_score(y_test, y_proba_log),
                roc_auc_score(y_test, y_proba_rf),
                roc_auc_score(y_test, y_proba_xgb)],
    'Accuracy': [log_model.score(X_test, y_test),
                 rf_model.score(X_test, y_test),
                 xgb_model.score(X_test, y_test)]
})

# Вывод таблицы
print(comparison)
```

### **Выводы:**

- **Accuracy (точность):**
  - Лучшую точность продемонстрировал **Random Forest (77.73%)**, что говорит о его способности правильно классифицировать большее количество клиентов.

#### 2. **Интерпретация результатов:**
- **Logistic Regression:**  
  Подходит для случаев, где требуется объяснимость модели, так как она предоставляет коэффициенты влияния признаков на вероятность оттока.
- **Random Forest:**  
  Показывает наивысшую точность, но его AUC-ROC ниже, чем у логистической регрессии. Эта модель может быть предпочтительнее, если важнее точность предсказания.
- **Gradient Boosting:**  
  Дает сбалансированные метрики и может быть полезен при сложных данных, но в данном случае уступает логистической регрессии по AUC-ROC.


### **Обоснование выбора модели:**
На основании метрик **лучшей моделью** для прогнозирования оттока является **Logistic Regression**, так как:
- Она показала **лучший AUC-ROC (0.8226)**, что указывает на высокую способность разделять классы `Churn` и `No Churn`.

#### Когда выбрать Random Forest:
Если бизнес требует максимально высокой точности классификации, можно использовать **Random Forest**. Однако при этом следует учитывать более низкий AUC-ROC.
