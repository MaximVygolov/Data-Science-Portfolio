**Описание данных:**

    Price — цена квартиры в рублях.
    Apartment type — тип жилья (вторичное или новостройка).
    Metro station — название ближайшей станции метро.
    Minutes to metro — время до метро в минутах.
    Region — регион, в котором находится квартира (например, Московская область).
    Number of rooms — количество комнат в квартире.
    Area — общая площадь квартиры в квадратных метрах (м²).
    Living area — жилая площадь квартиры в квадратных метрах (м²).
    Kitchen area — площадь кухни в квадратных метрах (м²).
    Floor — этаж, на котором расположена квартира.
    Number of floors — общее количество этажей в доме.
    Renovation — тип ремонта в квартире (например, косметический ремонт, евроремонт или без ремонта).
